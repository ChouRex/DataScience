```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 180)

y = np.cos(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()
```

```python
y_cos = np.cos(x)

plt.plot(x, y_cos)
plt.show()
```

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

T = np.arctan2(Y, X)
plt.scatter(X, Y, s = 75, c = T, alpha = 0.5)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()
```
