## #2嘗試通過添加紅色條形標籤重現上側的圖形

```python
import numpy as np
```

```python
# 配置12組bar
n = 12
X = np.arange(n)

#給定數學運算式

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

#指定上半部繪製區域, 給定 Bar 顏色, 邊界顏色

plt.bar(X, +Y1, facecolor = '#9999ff', edgecolor = 'white', alpha = 0.7)
#指定下半部繪製區域, 給定 Bar 顏色, 邊界顏色

plt.bar(X, -Y1, facecolor = 'red', edgecolor = 'white', alpha = 0.3)

# 設定繪圖圖示區間
for x, y in zip(X,Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha = 'center', va = 'bottom')

# 設定Y軸區間
plt.ylim(-1.25, +1.25)
plt.show()
```
