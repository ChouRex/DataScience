## Q1. 票價是否有異常值的現象？

Answer : 透過三種方法檢測，異常值大多落在票價為60以上的數值

```python
# 三倍標準差檢測
def outlier_z_score(data, times):
    mean = data.mean()
    std = data.std()
    z_score = [(y - mean) / std for y in data]
    return np.where(np.abs(z_score) > times)
```

```python
out_index = outlier_z_score(df_train["Fare"], 3)

print("三倍標準差的異常值為\n", df_train["Fare"].loc[out_index[0]])
```

```python
# 盒鬚圖解法
def outlier_iqr(data, times):
    quartile_1, quartile_3 = np.nanpercentile(data, [25, 75])
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * times)
    upper_bound = quartile_3 + (iqr * times)
    return np.where((data < lower_bound) | (data > upper_bound))
```

```python
out_index2 = outlier_iqr(df_train["Fare"], 1.5) 
df_train["Fare"].loc[out_index2]
```

## Q2. 你覺得找出的異常是真的異常? 你覺得需要做處理嗎?

Answer : 在盒鬚圖的檢測當中，可以看出大部分的異常值是屬於票價較高的部分。我想這在資料中不屬於異常值。這些票價較高的資料應該將他們拉出來做分群分析，而不是與票價較低的資料混在一起討論。就像是飛機票價一樣，坐頭等艙的人相對於整架飛機的人，他的票價絕對是被歸類在資料的異常值。反而我覺得該處理的是票價為0的資料，需要特別瞭解為什麼票價會=0，是因為有優惠卷嗎？還是是不知道艙位所以把票價設成0?這需要被瞭解。因此，在預測模型上面會考慮使用隨機森林，因其不會被異常值而影響。
