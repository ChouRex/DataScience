## Day_028 : 丟一個銅板，丟了100次，出現正面50次的機率有多大

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
# 定義二項分佈基本資訊

n = 100
p = 0.5

r = np.arange(0, 101)

probs = stats.binom.pmf(r, n, p)
print("擲骰子投擲到50次正面的機率 = ", probs[50])
plt.bar(r, probs)
plt.ylabel("P(X = x)")
plt.xlabel("x")
plt.title("(binomial(n = 100, p = 0.5)")
plt.show()
```

## Day_029_2. 大樂透中獎機率？

```python
#　定義超幾何分配的基本資訊

N = 49
K = 6
n = 6

r = np.arange(0, min(K, n) + 1)

# 計算 pmf

probs = stats.hypergeom.pmf(r, N, K, n) 
print("中頭獎的機率為 = ", probs[6])

plt.bar(r, probs)
plt.ylabel('P(X = x)')
plt.xlabel('x')
plt.title("pmf of Hypergeometric(N = 49, K = 6, N = 6)")
plt.show()
```
