## #1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?

#### Answer = "Tax" and "B"


```python
boston_data.boxplot(figsize = (12, 12))
```

```python
boston_data.median().between(300, 400) #透過between自行判斷
```

```python
# check
for name in boston_data.columns:
    if (400 > np.median(boston_data[name]) > 300):
        print(name)
```
